```python
#Problem 3

#part a:Use your function for N = 12, store it in variable named vander

def vandermonde (N):
    vec = np.arange (N)+1
    vander = vec[:,np.newaxis] ** np.arange(N)
    return vander
vander=vandermonde(12)
print(vander)    

#part b. Create a vector of all ones, of length 12 and call it x

x=np.ones(12,dtype=np.int64)
print(x)

b=np.dot(vander,x)  
print(b)

#part c:solve the linear system the naïve way, pretending like you don’t know x. Import numpy.linalg, invert V and multiply it by b

invert_V=np.linalg.inv(vander)
estimated_x1=invert_V @ b
print('Estimated x by inverted vander:\n',estimated_x1)
#Comment:some data points are exactly same to original X but there is some difference between estimated x and original x vector.It is not accurate.

#part d:solve the same linear system using solve.

estimated_x2=np.linalg.solve(vander,b)
print('Estimated x by solve function:\n',estimated_x2)

#comment:Here also some points are same to original x and some point throughing some difference between original x and estimated x.but solve funcion giving more accurate than invert method.
```
